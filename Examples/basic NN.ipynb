```python

x= np.random.normal(0,1, (100000, 3)) #Creamos un conjunto de datos de prueba
y_true = np.random.randint(0, 2, (100000,1)) #Creamos un vector de salida de ejemplo
y_test = np.random.random(size=(100000,1)) #Vector 

loss = tf.keras.losses.mean_squared_error(y_true, y_test)
```

### Contruimos la estructura de la red neuronal

<p>The optimization algorithm is key to minimize a function during the training.</p>
<p>this is applied on weights and learning rate.</p>
<p>The gradient descent uses constant learning rate. ADA algorithms use a variable learning rate.
ADAM: Adaptive Momentum estimation</p>
<p> Activation:</p>
<p>Backpropagation: passing error from front trough back of the NN.</p>
<p>Mini batch gradient descent: reduce the variance </p>

![Optimizer](images/1__osB82GKHBOT8k1idLqiqA.gif)


```python
model = tf.keras.models.Sequential(
    [
        tf.keras.layers.Dense(100, activation='relu', input_shape=(3,)),
        tf.keras.layers.Dense(500, activation='relu'),
        tf.keras.layers.Dense(1)
    ]
)
model.compile(optimizer='adam', loss='binary_crossentropy')

model.summary()
```

```python

model.fit(x,y_true,epochs=150,batch_size=100)
```

```python
history = model.history
loss =  history.history["loss"]
epochs = range(1, len(loss) + 1)


plt.plot(epochs, loss,  label='Training loss')
plt.title('Training  loss')
plt.legend()

plt.show()
```
